```python
import pandas as pd
df = pd.read_csv('/content/diabetes.csv')
```

```python
df.info()
df.describe()
```

```python
df.head()
```

```python
#Null Value assiging
df.Pregnancies[0] = None
df.Pregnancies[10] = None
df.Pregnancies[100] = None
df.Pregnancies[200] = None
df.Pregnancies[111] = None
df.BloodPressure[0] = None
df.BloodPressure[11] = None
df.BloodPressure[111] = None
```

```python
df.head()
```

```python
#Show the total number of null contains  each column
df.isna().sum()
```

```python
#drop the the whole row which contains null, but no change occurs in original dataset
df.dropna()
```

```python
df
```

```python
#fill null with mean value, and change occurs in original dataset because of (inplace=True)
df.fillna(df.Pregnancies.mean(),inplace=True)
```
