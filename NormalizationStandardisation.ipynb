# **MinMax + Standardization**

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
scaling1 = MinMaxScaler()#Normaliztion(convert between 0-1)
scaling2 = StandardScaler()#standardization(-1 to 1)
```

```python
df = pd.read_csv('/content/fish.csv')
```

```python
df.head()
```

```python
scaling1.fit_transform(df.iloc[:,1:2])
```

```python
scaling2.fit_transform(df.iloc[:,1:2])
```

# **Z_Score**

```python
import scipy.stats as stats
df.iloc[:,3:] = stats.zscore(df.iloc[:,3:] )
```
